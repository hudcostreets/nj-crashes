```python
ints = [ 'Total Killed', 'Total Injured', 'Pedestrians Killed', 'Pedestrians Injured', 'Total Vehicles Involved', ]
floats = [ 'Latitude', 'Longitude', 'MilePost', ]
bools = [ 'Alcohol Involved', 'HazMat Involved', ]

nj2 = nj.copy()

nj2['Date'] = to_dt(nj2['Crash Date'] + ' ' + nj2['Crash Time'])
years = nj2['Year']
nj2 = nj2.drop(columns=['Year', 'Crash Time', 'Crash Date', 'Crash Day Of Week'])

for k in floats:
    nj2[k] = nj2[k].replace('', nan).astype(float)

for k in bools:
    nj2[k] = nj2[k].replace('Y', True).replace('N', False).astype(bool)

nj2 = nj2.astype(dict(**{ k: int for k in ints },))

nj2
```

```python
months = nj2.groupby([ nj2.Date.dt.year.rename('year'), nj2.Date.dt.month.rename('month') ])[['Total Killed', 'Total Injured']].sum(0).reset_index()
months['month'] = months.apply(lambda r: to_dt('%d-%02d' % (r.year, r.month)), axis=1)
months = months.drop(columns='year').set_index('month')
months
```

```python
gridcolor = '#ccc'
def month_totals_plot(s, typ):
    fig = px.bar(
        s,
        color_discrete_sequence=['black'],
        labels={ 'variable': '', 'value': 'Injuries', 'month': 'Month', },
    ).update_layout(
        plot_bgcolor='white',
        paper_bgcolor='white',
        title=f'NJ Traffic Crash {typ} Per Month',
        title_x=0.5,
        showlegend=False,
    ).update_xaxes(
        gridcolor=gridcolor,
    ).update_yaxes(
        gridcolor=gridcolor,
    )
    fig.write_image(f'{typ.lower()}_per_month.png', width=1200, height=600)
    return fig
```

```python
month_totals_plot(months['Total Injured'], 'Injuries')
```

```python
month_totals_plot(months['Total Killed'], 'Deaths')
```
