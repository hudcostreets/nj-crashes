### Cross-reference 2019 data above vs. "Uniform Crime Report"
https://nj.gov/njsp/ucr/uniform-crime-reports.shtml

```python
def ucr_murder_total(path):
    sheets = pd.read_excel(path, None)
    county_murder_totals = {}

    for sheet_name, sheet in sheets.items():
        if sheet_name == 'DocumentMap':
            continue

        sheet = sheet.copy()
        sheet.columns = sheet.iloc[2]
        murder_counts = sheet[sheet.Agency == 'Number of Offenses'].Murder
        if murder_counts.empty:
            total = precinct_total = 0
        else:
            total = murder_counts.iloc[-1]
            precinct_total = murder_counts.iloc[:-1].sum()
        assert precinct_total == total, f'{sheet_name}: {precinct_total} != {total}'
        county_murder_totals[sheet_name] = total

    county_murder_totals['Total'] = sum(county_murder_totals.values())
    return county_murder_totals
```

```python
ucr_xlsx_start = 2018
ucr_end = 2024
prefixes = { 2023: '20250416_' }
ucr = DF([
    dict(year=year, county=county, murders=murders)
    for year in range(ucr_xlsx_start, ucr_end)
    for county, murders in ucr_murder_total(f'{prefixes.get(year, "")}{year}_Uniform_Crime_Report.xlsx').items()
])
ucr
```

```python
tot_mask = ucr.county == "Total"
ucr_nj = ucr[tot_mask].set_index('year').murders
sp_mask = ucr.county == "State Police"
ucr_sp = ucr[sp_mask].set_index('year').murders
ucrs = ucr[~tot_mask & ~sp_mask].set_index(['year', 'county']).murders
ucr_nj
```

### Compute ≥2020 NJ murder total

```python
murders_per_year = d.Murder.copy()
for y in range(2020, ucr_end):
    murders_per_year.loc[y] = ucr_nj.loc[y]
murders_per_year
```

```python
save(
    px.bar(
        murders_per_year,
        labels={'value': ''},
        color_discrete_sequence=['black'],
    ),
    title="NJ: homicides per year",
    name='homicides-per-year',
    legend=False,
    xaxis_dtick=5,
);
```
